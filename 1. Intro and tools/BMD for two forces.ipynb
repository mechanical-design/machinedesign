```python
from matplotlib import pyplot as plt
```

```python
#BMD

coorC = [0,0]
coorA = [400,0]
coorB = [1000,0]
coorD = [1400,0]

Cy=-1689
Ay=6364
By=-10000
Dy=5325

cd=1400
ca=400
cb=1000

# Moment is area under the graph of shear force
y = [0,ca,cb,cd]
M = [0,Cy*ca/1000,(Cy*ca/1000)+((Cy+Ay)*(cb-ca)/1000),0]

fig, ax2 = plt.subplots()

ax2.plot(y,M)
ax2.hlines(0,0,cd,colors='black')
ax2.fill_between(y, M, alpha=0.3) # Added fill_between
ax2.set_title('Bending Diagram (BMD)')
ax2.set_xlabel('Distance along the beam,x (mm)')
ax2.set_ylabel('Bending Moment,M (Nmm)')

ax2.text(coorA[0],coorA[1]+10,f'  A')
ax2.text(coorB[0],coorB[1]+10,f'  B')
ax2.text(coorC[0],coorC[1]+10,f'  C')
ax2.text(coorD[0],coorD[1],f'  D')

ax2.text(coorB[0],M[2],f' Mmax={M[2]}Nm')

plt.tight_layout()
plt.show()
```
