```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

Draw Shear Force Diagram(SFD) and Bending Moment Diagram (BMD) for simply supported beam, one concentrated force.

```python
# SFD for a concentrated load at C
# Distance in mm

coorA = [0,0]
coorB = [800,0]
coorC = [300,0]

ab=800
ac=300

# Forces in N
Fa=500
Fc=1000

# Shear force is plotted manually from left to right
y = [0,0,ac,ac,ab,ab]
V = [0,Fa,Fa,Fa-Fc,Fa-Fc,0]

fig, ax1 = plt.subplots()
ax1.set_yticks([0])
ax1.plot(y,V)
ax1.hlines(0,0,ab,colors='black')
ax1.fill_between(y, V, alpha=0.3) # Added fill_between
ax1.set_title('Shear Force Diagram (SFD)')
ax1.set_xlabel('Distance along the beam,l (mm)')
ax1.set_ylabel('Shear Force,V (N)')

# Label
ax1.text(coorA[0],coorA[1]+50,f'  A')
ax1.text(coorB[0],coorB[1]+50,f'  B')
ax1.text(coorC[0],coorC[1]+50,f'  C')
ax1.text(coorA[0],V[1]+10,f'        {V[1]}N')
ax1.text(coorC[0],V[3]+10,f'                  {V[3]}N')
# plt.tight_layout()
plt.show()
```

```python
# BMD

# Bending moment is area under SFD from left to right
y=[0,ac,ab]
M=[0,Fa*ac/1000,0]

fig, ax2 = plt.subplots()
ax2.set_yticks([0])
ax2.plot(y,M)
ax2.hlines(0,0,ab,colors='black')
ax2.fill_between(y, M, alpha=0.3) # Added fill_between
ax2.set_title('Bending Diagram (BMD)')
ax2.set_xlabel('Distance along the beam,l (mm)')
ax2.set_ylabel('Bending Moment,M (Nm)')

ax2.text(coorA[0],coorA[1]+5,f'  A')
ax2.text(coorB[0],coorB[1]+5,f'B')
ax2.text(coorC[0],coorC[1]+5,f'C')

ax2.text(coorC[0],M[1],f' Mmax={M[1]}Nm')

plt.tight_layout()
plt.show()
```
