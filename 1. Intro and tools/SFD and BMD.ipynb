```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ay=286.43
Cy=381.9
By=137.89
ab=400
ac=100
y = np.linspace(0,ab,100)

V = np.zeros_like(y)
for i, xi in enumerate(y):
    if xi < ac:
        V[i] = Ay
    else:
        V[i] = Ay - Cy

M = np.zeros_like(y) # Create Moment for each distance
for i, xi in enumerate(y):
    if xi <= ac:
        M[i] = Ay * xi/1000
    else:
        M[i] = Ay * xi/1000 - Cy * (xi - ac)/1000

# plt.figure(figsize=(6,12))

fig, (ax1,ax2) = plt.subplots(2,1)
ax1.plot(y,V)
ax1.fill_between(y, V, alpha=0.3) # Added fill_between
ax1.set_title('Shear Force Diagram (SFD)')

ax2.plot(y,M)
ax2.fill_between(y, M, alpha=0.3) # Added fill_between
ax2.set_title('Bending Diagram (BFD)')

plt.tight_layout()
plt.show()
```
