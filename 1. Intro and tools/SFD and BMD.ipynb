```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
# SFD for a concentrated load at C
# Distance in mm
ab=800
ac=300

# Forces in N
Fa=500
Fc=1000

y = [0,0,ac,ac,ab,ab]
V = [0,Fa,Fa,Fa-Fc,Fa-Fc,0]

fig, ax1 = plt.subplots()
# SFD
ax1.plot(y,V)
ax1.hlines(0,0,ab,colors='black')
ax1.fill_between(y, V, alpha=0.3) # Added fill_between
ax1.set_title('Shear Force Diagram (SFD)')

# plt.tight_layout()
plt.show()
```

```python
# Draw SFD and BMD - General

# State once
coorA = [0,0]
coorB = [330,0]
coorC = [100,0]
coorD = [250,0]

# Force with load at C
Ay=286.43
Cy=381.9
By=137.89

# Distance in mm
ab=400
ac=100
y = np.linspace(0,ab,100) # Create distance variables
V = np.zeros_like(y) # Create Shear Force for each distance
for i, xi in enumerate(y):
    if xi < ac:
        V[i] = Ay
    else:
        V[i] = Ay - Cy

M = np.zeros_like(y) # Create Moment for each distance
for i, xi in enumerate(y):
    if xi <= ac:
        M[i] = Ay * xi/1000
    else:
        M[i] = Ay * xi/1000 - Cy * (xi - ac)/1000



fig, (ax1,ax2) = plt.subplots(2,1)
# SFD
ax1.plot(y,V)
ax1.hlines(0,0,400,colors='black')
ax1.fill_between(y, V, alpha=0.3) # Added fill_between
ax1.set_title('Shear Force Diagram (SFD)')
ax1.set_ylabel('V(N)')

#BMD
ax2.plot(y,M)
ax2.fill_between(y, M, alpha=0.3) # Added fill_between
ax2.set_title('Bending Diagram (BMD)')
ax2.set_ylabel('M(Nmm)')

#Add the vertical lines at the start (y=0) and end (y=ab)
ax1.vlines(x=0, ymin=0, ymax=V[1])      # Start
ax1.vlines(x=ab, ymin=0, ymax=V[-1])   # End


ax2.text(coorA[0],coorA[1],f'  A')
ax2.text(coorB[0],coorB[1],f'  B')
ax2.text(coorC[0],coorC[1],f'  C')
ax2.text(coorD[0],coorD[1],f'  D')

ax2.set_ylim(0,30)

plt.tight_layout()
plt.show()
```
