Relation between Force at bolt and proof load

```python
# Required
import matplotlib.pyplot as plt
import numpy as np

# Parameters
C = 0.3
PLoad = 1000
Nbolt = 1

Sp = 380E+6 # From Table
At = 58E-6 # From Table
Fp=Sp*At
Fi = 0.75*Fp #Reusable joint
```

```python
# Equation
Fbolt = C*(PLoad/Nbolt)+Fi
print(f'Fbolt= {Fbolt:.3e}N')
```

```python
PLoad = np.arange(0,11000,1000)
Fbolt = C*(PLoad/Nbolt)+Fi

Fproof = Sp*At

fig, ax = plt.subplots()
ax.plot(PLoad, Fbolt)
ax.hlines(Fproof,0,10000)
ax.hlines(Fi,0,10000)

ax.set(xlabel='Load, P', ylabel='Force at Bolt, Fbolt',
       title='Relation between Total force at Bolt')
ax.grid()
plt.show()
```
