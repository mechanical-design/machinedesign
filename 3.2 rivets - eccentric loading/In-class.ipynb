Find centroid of the bolt group with coordinates (0,0), (150,0), (150,120)
and (0,120) mm.

```python
A = [0,0]
B = [150,0]
C = [150,120]
D = [0,120]

x_bar = (A[0]+B[0]+C[0]+D[0])/4
y_bar = (A[1]+B[1]+C[1]+D[1])/4
print(f"Centroid (x_bar, y_bar) = ({x_bar:.2f}, {y_bar:.2f}) mm")
```

Find Shear Stress if load of 5kN is applied to a bolt with diameter of
16mm.

```python
d=0.016
P=5000
Area = pi*d**2/4
shear_stress = P/Area
print(f"Shear stress = {shear_stress/1e6:.2f} MPa")
```

A bracket in the form of plate is fitted to a column by a number of rivets.
The riveted joint is under an inclined eccentric force. The plate is 15 mm
thick and the diameter of rivet is 20 mm. Determine the bearing stress if
the resultant shear force is 20 kN.

```python
P=20e3
d=20e-3
t=15e-3
bearing_stress = P/(d*t)
print(f"Bearing stress = {bearing_stress:.3e} Pa")
```
