## Question 1

A plastic flat belt 60 mm wide and 0.5 mm thick transmits 10 kW. The input pulley has a diameter of 300 mm, it rotates at 2800 rpm, and the output pulley speed is 1600 rpm; the pulleys are 700 mm apart, the coefficient of friction is 0.2, and belt weight is 25 kN/m3. It is assumed that the driver is a high torque motor, and the driven machine is under a medium shock load, $K_s=1.5$.

i - Find torque at smaller pulley 

ii - The contact angle 

iii - The maximum tension and stress in the belt

```python
width = 0.06
thickness = 0.0005
Power = 10000
d_input = 0.3
n_input = 2800
n_output =1600
c_distance = 0.7
friction =0.2
belt_weight = 25000
Ks = 1.5
```

i - Find torque at smaller pulley 

```python
T_driver=Power*60/(2*pi*n_input)
print(f'Torque on driver pulley: {T_driver:.2f} Nm')
```

ii - The contact angle 

$$sin \alpha = \dfrac{r_2 - r_1}{c}$$

$$n_1 d_1 = n_2 d_2$$

```python
d_output = (n_input/n_output)*d_input
print(f'Diameter of output pulley: {d_output:.4f} m')
```

```python
r2 = d_output/2
r1 = d_input/2
alpha = asin((r2 - r1)/c_distance)
print(f'Alpha ={alpha:.4f} rad')
print(f'Convert to degrees, Alpha ={degrees(alpha):.2f} degrees')
```

```python
contact_angle_driver = round(pi - (2*alpha),4)
print(f'Contact angle for driver pulley ={contact_angle_driver:.4f} rad')
print(f'Convert to degrees, Contact angle for driver pulley ={degrees(contact_angle_driver):.2f} degrees')
```

```python
contact_angle_driven = pi + (2*alpha)
print(f'Contact angle for driven pulley ={contact_angle_driven:.4f} rad')
print(f'Convert to degrees, Contact angle for driven pulley ={degrees(contact_angle_driven):.2f} degrees')
```

iii - Centrifugal tensions in the belt, Fc

$$F_c = \dfrac{w}{g}v^2$$

```python
w = belt_weight * width * thickness
print(f'Weight of the belt per unit length: {w:.2f} N/m')
```

```python
velocity = round((pi*d_input*n_input)/60,2)
print(f'Velocity of the belt: {velocity:.2f} m/s')
```

```python
Fc=round((w/9.81)*velocity**2,2)
print(f'Centifugal force acting on the belt: {Fc:.2f} N')
```

$$\dfrac{F_1-F_c}{F_2-F_c} = e^{f \phi}$$

```python
e_f_phi = e**(friction*contact_angle_driver)
print(f'efi for driver pulley: {e_f_phi:.4f}')
```

```python
gamma = round(e**(friction*contact_angle_driver),4)
print(f'Value of gamma: {gamma:.4f}')
F1=Fc+(gamma/(gamma-1))*T_driver/(d_input/2)
print(f'Tension on the tight side of the belt on driver pulley: {F1:.2f} N')
```

Stress on belt

```python
sigma = F1/(width*thickness)
print(f'Stress in the belt: {sigma/1e6:.2f} MPa')
```
