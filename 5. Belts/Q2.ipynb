Q2

V belt drive has a 200mm diameter small sheave with 170° contact angle, 38° included angle, coefficient of friction of 0.15, 1600rpm driver speed, belt weight 8N/m and a tight side tension of 3kN. Determine the power capacity of the drive.

```python
d_driver = 0.2
contact_angle = 170/180 * pi
print("Contact Angle (radians): ", contact_angle)
included_angle = 38/2
friction = 0.15
n_driver = 1600
belt_weight = 8
F1=3000
```

```python
velocity = round((pi * d_driver * n_driver)/60,2)
print("Velocity (m/s): ", velocity)
```

```python
Fc = round((belt_weight/9.81)*velocity**2,2)
print(f'Centrifugal Force (N): {Fc}N')
```

```python
print(friction, contact_angle, included_angle)
f_phi=round((friction*contact_angle/sin(radians(included_angle))),3)
print(f'f_phi: {f_phi}')
```

```python
F1_minus_Fc = F1 - Fc
print(f'F1 - Fc: {F1_minus_Fc}N')
```

```python
efi=round((e**(f_phi)),4)
print(f'efi: {efi}')
F2=round((F1_minus_Fc/efi)+(Fc),2)
print(f'F2: {F2}N')
```

```python
Torque_driver = round((F1-F2)*(d_driver/2),2)
print(f'Torque on Driver (Nm): {Torque_driver}Nm')
```

```python
Power=round((2*pi*n_driver*Torque_driver)/60,2)
print(f'Power Transmitted (W): {Power}W')
Power_kW=round(Power/1000,2)
print(f'Power Transmitted (kW): {Power_kW}kW')
```
