Q3

A v belt drives has an included angle of 38°, belt weight 3N/m, belt cross sectional are of 145mm², coefficient of friction of 0.25, r₁=150mm, Ø=160, n₁=3000rpm and F₂=800N. Calculate

i - Maximum power transmitted

ii - Maximum stress in belt

```python
included_angle = 38/2
# belt_weight = 3
rho = 2100  # kg/m^3
Area = 145e-6
friction = 0.25
r_driver = 0.15
d_driver = r_driver*2
contact_angle = round(160/180 * pi,4)
print(f'Contact Angle (rad): {contact_angle}rad')
n_driver = 3000
F2=800
```

```python
velocity = round((pi * r_driver*2 * n_driver)/60,2)
print("Velocity (m/s): ", velocity)
```

```python
Fc = round(rho*Area*velocity**2,0)
print(f'Centrifugal Force (N): {Fc}N')
```

```python
print(friction, contact_angle, included_angle)
f_phi=round((friction*contact_angle/sin(radians(included_angle))),3)
print(f'f_phi: {f_phi}')
```

```python
efi=round((e**(f_phi)),4)
print(f'efi: {efi}')
```

```python
F2_minus_Fc = F2 - Fc
print(f'F2 - Fc: {F2_minus_Fc}N')
```

```python
F1 = round((F2_minus_Fc*efi)+(Fc),2)
print(f'F1: {F1}N')
```

```python
Torque_driver = (F1-F2)*r_driver
print(f'Driver Torque (Nm): {Torque_driver}Nm')
```

```python
Power_driver = round((Torque_driver * (n_driver * (2 * pi) / 60)),2)
print(f'Driver Power (W): {Power_driver}W')
print(f'Driver Power (kW): {Power_driver/1000}kW')
```

```python
sigma = round(F1/Area,2)
print(f'Belt Stress (Pa): {sigma}Pa')
print(f'Belt Stress (MPa): {sigma/1e6}MPa')
```
