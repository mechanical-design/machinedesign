Q4

In a horizontal V belt drive for a centrifugal blower, the blower is belt driven at 600 rpm by a 15 kW, 1750rpm electric motor. The groove angle of the pulley is 35Â°. The center distance is twice the diameter of the larger pulley and velocity of the belt is 20 m/s. The density of the belt material is 1500kg/m3 and maximum allowable stress is 4MPa. The coefficient of friction of motor pulley is 0.5. Determine

the diameter of the blower pulley

the belt length

the size of the belt in mm2


```python
Power=15000
n_driver=1750
n_driven=600
half_included_angle=35/2
friction=0.5
velocity=20
rho=1500
sigma=4e6
```

```python
d_driver = round((60 * velocity) / (pi * n_driver),3)
print(f'Driver Pulley Diameter (m): {d_driver}m')
```

$n_1 d_1 = n_2 d_2$

```python
d_driven = round((n_driver / n_driven) * d_driver,3)
print(f'Driven Pulley Diameter (m): {d_driven}m')
```

```python
center_distance = 2*d_driven
print(f'Center Distance (m): {center_distance}m')
```

```python
belt_length = round(2*center_distance + (pi/2)*(d_driver + d_driven) + ((d_driven - d_driver)**2)/(4*center_distance),3)
print(f'Belt Length (m): {belt_length}m')
```

```python
FcA = round(rho*velocity**2,0)
print(f'Let Centrifugal Force, Fc={FcA} A')
```

```python
Torque_driver = round((Power*60/(2*pi*n_driver)),2)
print(f'Driver Pulley Torque (Nm): {Torque_driver}Nm')
```

```python
r1=d_driver/2
print(f'Driver Pulley Radius (m): {r1}m')
```

```python
F1_minus_F2 = round((Torque_driver)/r1,2)
print(f'Tension Difference, F1 - F2 (N): {F1_minus_F2}N')
```

```python
r1 = d_driver / 2
r2 = d_driven / 2
print(f'Driver Pulley Radius (m): {r1}m')
print(f'Driven Pulley Radius (m): {r2}m')
alpha = round(asin((r2 - r1) / center_distance),3)
print(f'Alpha {alpha}rad')
phi = round(pi - (2*alpha),3)
print(f'Phi {phi}rad')
```

```python
efi=round(e**(friction*phi/sin(radians(half_included_angle))),1)
print(f'Right hand side {efi}')
```
