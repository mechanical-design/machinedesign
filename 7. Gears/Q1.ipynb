```python
# Required
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, tan, pi, atan, asin, degrees, radians
```

The gears shown in Figure below have a module of 6 mm and a 20Â° pressure angle. Determine, and show on a free-body diagram,

1. The tangential and radial loads on each gear.
2. The reactions on shaft C.

Given: Driving gear 1 transmits 80 kW at 1600 rpm through the idler to gear 3 mounted on shaft C.

```python
Power = 80000
n1=1600
module = 6
pressureAngle=radians(20)
N1=18
N2=30
N3=50
```

```python
d1=module*N1
d2=module*N2
d3=module*N3
print(f'd1={d1}')
print(f'd2={d2}')
print(f'd3={d3}')
r1=d1/2
r2=d2/2
r3=d3/2
print(f'r1={r1}')
print(f'r2={r2}')
print(f'r3={r3}')
```

Find rotational speed for all gears

$$\frac{n_1}{n_2}=\frac{N_2}{N_1}$$
$$\frac{1600}{n_2}=\frac{30}{18}$$

```python
n2=n1*N1/N2
print(f'n2={n2:.0f}rpm')
```

```python
# Draw Gear
fig, ax = plt.subplots()

circle1 = plt.Circle((0, 0), r1, color='blue', fill=False)  
circle2 = plt.Circle((r1+r2, 0), r2, color='green', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-80, 250)
ax.set_ylim(-120, 120)

# ax.set(title='Gear 1 and 2')
ax.text(0,r1+20,f'Gear 1')
ax.text(r1+r2,r2+20,f'Gear 2')
ax.text(0,-r1-20,f'd1={d1}mm')
ax.text(r1+r2+20,-r2-20,f'd2={d2}mm')
plt.show()
```

```python
# FBD between Gear 1 and 2
fig, ax = plt.subplots()

offsetx=40
offsety=10

circle1 = plt.Circle((0, 0), r1, color='blue', fill=False)  
circle2 = plt.Circle((r1+r2+offsetx, 0), r2, color='green', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-80, 280)
ax.set_ylim(-120, 120)

ax.set(title='FBD between Gear 1 and 2')
ax.text(0,r1+10,f'Gear 1')
ax.text(r1+r2+offsetx,r2+10,f'Gear 2')
ax.text(0,-r1-offsety,f'n1={n1} rpm')
ax.text(r1+r2+offsetx,-r2-offsety,f'n2={n2:.0f} rpm')

ax.arrow(r1, 0, 0,-40,width=2,color='r')
ax.text(r1,-40,f'Ft1')
ax.arrow(r1+offsetx, 0, 0,40,width=2,color='r')
ax.text(r1+offsetx,40,f'Ft2')

ax.arrow(r1, 0, -20,0,width=2,color='r')
ax.text(r1-offsetx,0,f'Fr1')
ax.arrow(r1+offsetx, 0, 20,0,width=2,color='r')
ax.text(r1+offsetx+20,0,f'Fr2')

ax.annotate("", xy=(r1, 30), xytext=(r1, -30),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
ax.annotate("", xy=(r1+offsetx, 30), xytext=(r1+offsetx, -30),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2",
                            color="green", lw=2))
plt.show()
```

Torque

$$P=\omega T$$
$$P=\frac{2\pi n_1}{60} T_1$$
$$80000=\frac{2\pi (1600)}{60} T_1$$
<!-- $$T=7.16 N.m$$ -->

```python
T_1=round(Power*60/(2*pi*n1),2)
print(f'Torque={T_1} Nm')
```

Find radius for Gear 1
$$r_1=\frac{0.108}{2}$$
$$r_1=0.054 m$$

Use torque equation

$$T=F_{t1}\times r_1$$
$$477.46=F_{t1}(0.054)$$

```python
r1=d1/2
print(f'r1={r1} mm')
Ft1=round(T_1/(r1/1000),2)
print(f'Ft1={Ft1} N')
```

Radial force

$$F_{r1}=F_{t1} \tan(20)$$
$$=(8841.85)\tan(20)$$

```python
Fr1=round(Ft1*tan(pressureAngle),2)
print(f'Fr1={Fr1}N')
```

Gear 2 and 3

$$\frac{n_2}{n_3}=\frac{N_3}{N_2}$$
$$\frac{960}{n_3}=\frac{50}{30}$$

```python
n3=n2*N2/N3
print(f'n3={n3:.0f}rpm')
```

```python
# Draw Gear
fig, ax = plt.subplots()

circle1 = plt.Circle((0, 0), r1, color='blue', fill=False)  
circle2 = plt.Circle((r1+r2, 0), r2, color='green', fill=False)  
circle3 = plt.Circle((r1+2*r2+r3, 0), r3, color='brown', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-150, 550)
ax.set_ylim(-200, 200)

# ax.set(title='Gear 1 and 2')
ax.text(0,r1+20,f'Gear 1')
ax.text(r1+r2,r2+20,f'Gear 2')
ax.text(r1+2*r2+r3,r3+20,f'Gear 3')
ax.text(0,-r1-20,f'd1={d1}mm')
ax.text(r1+r2,-r2-20,f'd2={d2}mm')
ax.text(r1+2*r2+r3,-r3-20,f'd3={d3}mm')
plt.show()
```

```python
# FBD between Gear 3 and 4
fig, ax = plt.subplots()

offsetx=40
offsety=10

circle1 = plt.Circle((0, 0), r2, color='green', fill=False)  
circle2 = plt.Circle((r2+r3+offsetx, 0), r3, color='brown', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-150, 450)
ax.set_ylim(-200, 200)

ax.set(title='FBD between Gear 2 and 3')
ax.text(0,r2+10,f'Gear 2')
ax.text(r2+r3+offsetx,r3+10,f'Gear 3')
ax.text(0,-r2-offsety,f'n2={n2} rpm')
ax.text(r2+r3+offsetx,-r3-offsety,f'n3={n3:.0f} rpm')

# Left Gear 2 Forces
ax.arrow(-r2, 0, 0,40,width=2,color='r')
ax.text(-r2,40,f'Ft2')
ax.arrow(-r2, 0, 20,0,width=2,color='r')
ax.text(-r2+offsetx,0,f'Fr2')

# Right Gear 2 Forces
ax.arrow(r2, 0, 0,40,width=2,color='r')
ax.text(r2,40,f'Ft2')
ax.arrow(r2, 0, -20,0,width=2,color='r')
ax.text(r2-offsetx,0,f'Fr2')

ax.arrow(r2+offsetx, 0, 0,-40,width=2,color='r')
ax.text(r2+offsetx,-40,f'Ft3')
ax.arrow(r2+offsetx, 0, 20,0,width=2,color='r')
ax.text(r2+offsetx+20,0,f'Fr3')


ax.annotate("", xy=(r2, 30), xytext=(r2, -30),
            arrowprops=dict(arrowstyle="<-",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
ax.annotate("", xy=(r2+offsetx, -30), xytext=(r2+offsetx, 30),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
plt.show()
```
