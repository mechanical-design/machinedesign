# Q4(b) - Gear Trains

In gear train shown in Figure Q4(b), gear 1 is the inlet of the gear train and transfers 1.8kW of power at 2400 rpm. The module of all gears in the train drive is 4 mm and the pressure angle is 20Â°. Diameter for gear 1, 2, 3 and 4 are 32mm, 60mm, 80mm and 100mm respectively.

i) Determine the number of teeth for gears 1, 2, 3 and 4.

ii) Determine the tangential and radial forces between gears 1 and 2 and between
gears 3 and 4.

iii) If the diameter of gear 4 is increased to 120mm, explain how it affects the torque of gear 4

```python
# Required
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos, tan, pi, atan, asin, degrees, radians
```

```python
Power = 1800
n1=2400
module = 4
pressureAngle=radians(20)
d1=32
d2=60
d3=80
d4=100
# print('Speed ratio by 5:2')
```

Module
$$m=\frac{d}{N}$$

```python
N1=d1/module
N2=d2/module
N3=d3/module
N4=d4/module
print(f'N1={N1}')
print(f'N2={N2}')
print(f'N3={N3}')
print(f'N4={N4}')
```

```python
r1=d1/2
r2=d2/2
r3=d3/2
r4=d4/2
print(f'r1={r1}')
print(f'r2={r2}')
print(f'r3={r3}')
print(f'r4={r4}')
```

ii) Determine tangential and radial forces between Gear 1 and 2 and between gear 3 and 4.

Find rotational speed for all gears

$$\frac{n_1}{n_2}=\frac{N_2}{N_1}$$
$$\frac{2400}{n_2}=\frac{15}{8}$$

```python
n2=n1*N1/N2
print(f'n2={n2:.0f}rpm')
```

```python
# Draw Gear
fig, ax = plt.subplots()

circle1 = plt.Circle((0, 0), r1, color='blue', fill=False)  
circle2 = plt.Circle((r1+r2, 0), r2, color='green', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-20, 80)
ax.set_ylim(-40, 40)

# ax.set(title='Gear 1 and 2')
ax.text(0,r1+5,f'Gear 1')
ax.text(r1+r2,r2+5,f'Gear 2')
ax.text(0,-r1-5,f'd1={d1}mm')
ax.text(r1+r2+5,-r2-5,f'd2={d2}mm')
plt.show()
```

```python
print('Rotation speed for gear 2 and 3 are same')
n3=n2
```

$$n_2=n_3=1200rpm$$
$$\frac{n_3}{n_4}=\frac{N_4}{N_3}$$
$$\frac{1200}{n_4}=\frac{400}{320}$$

```python
n4=n3*N3/N4
print(f'n4={n4:.0f}rpm')
```

Draw FBD for Gear 1 and 2

```python
# FBD between Gear 1 and 2
fig, ax = plt.subplots()

offsetx=10
offsety=5

circle1 = plt.Circle((0, 0), r1, color='blue', fill=False)  
circle2 = plt.Circle((r1+r2+offsetx, 0), r2, color='green', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-20, 100)
ax.set_ylim(-40, 40)

ax.set(title='FBD between Gear 1 and 2')
ax.text(0-offsetx,r1+1,f'Gear 1')
ax.text(r1+r2+offsetx,r2+1,f'Gear 2')
ax.text(0,-r1-offsety,f'n1={n1} rpm')
ax.text(r1+r2+offsetx,-r2-offsety,f'n2={n2:.0f} rpm')

ax.arrow(r1, 0, 0,20,width=0.5,color='r')
ax.text(r1,20+offsety,f'Ft1')
ax.arrow(r1, 0, -10,0,width=0.5,color='r')
ax.text(r1-18,0,f'Fr1')

ax.arrow(r1+offsetx, 0, 0,-20,width=0.5,color='r')
ax.text(r1+offsetx,-20-offsety,f'Ft2')
ax.arrow(r1+offsetx, 0, 10,0,width=0.5,color='r')
ax.text(r1+offsetx+15,0,f'Fr2')

ax.annotate("", xy=(r1, -5), xytext=(r1, 5),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=-0.2",
                            color="green", lw=2))
ax.annotate("", xy=(r1+offsetx, -5), xytext=(r1+offsetx, 5),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
plt.show()
```

Torque

$$P=\omega T$$
$$P=\frac{2\pi n_1}{60} T$$
$$1800=\frac{2\pi 2400}{60} T$$
$$T=7.16 N.m$$

```python
Torque=round(Power*60/(2*pi*n1),2)
print(f'Torque={Torque} Nm')
```

Find radius for Gear 1
$$r_1=\frac{0.03}{2}$$
$$r_1=0.015 m$$


```python
r1=d1/2/1000
print(f'r1={r1} m')
```

Use torque equation

$$T=F_{t1}\times r_1$$
$$7.16=F_{t1}(0.016)$$

```python
Ft1=round(Torque/r1,2)
print(f'Tangential Force, Ft1={Ft1} N')
```

Radial force

$$F_{r1}=F_{t1} \tan(20)$$
$$=(477.33)\tan(20)$$

```python
Fr1=round(Ft1*tan(pressureAngle),2)
print(f'Fr1={Fr1}N')
```

$$F_{t1}=F_{t2}$$
$$F_{r1}=F_{r2}$$

```python
# Draw Gear
fig, ax = plt.subplots()

offsetx=10
offsety=5

circle1 = plt.Circle((0, 0), r3, color='green', fill=False)  
circle2 = plt.Circle((r3+r4, 0), r4, color='brown', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-60, 150)
ax.set_ylim(-80, 80)

# ax.set(title='Gear 1 and 2')
ax.text(-offsetx,r3+offsety,f'Gear 3')
ax.text(-offsetx+r3+r4,r4+offsety,f'Gear 4')
ax.text(-offsetx,-r3-offsety,f'd3={d3}mm')
ax.text(r3+r4,-r4-offsety,f'd4={d4}mm')
plt.show()
```

Draw FBD for Gear 3 and 4

```python
# FBD between Gear 3 and 4
fig, ax = plt.subplots()

offsetx=40
offsety=10

circle1 = plt.Circle((0, 0), r2, color='green', fill=False)  
circle2 = plt.Circle((r2+r3+offsetx, 0), r3, color='brown', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-50, 160)
ax.set_ylim(-60, 60)

ax.set(title='FBD between Gear 3 and Gear 4')
ax.text(-5,r2+offsety,f'Gear 3')
ax.text(r3+r4,r3+offsety,f'Gear 4')
ax.text(-offsetx,-r2-offsety,f'n3={n3:.0f} rpm')
ax.text(r2+r3+offsetx,-r3-offsety,f'n4={n4:.0f} rpm')

ax.arrow(r2, 0, 0,-40,width=1,color='r')
ax.text(r2,-45,f'Ft3')
ax.arrow(r2, 0, -20,0,width=1,color='r')
ax.text(r2-offsetx,0,f'Fr3')

ax.arrow(r2+offsetx, 0, 0,40,width=1,color='r')
ax.text(r2+offsetx,45,f'Ft4')
ax.arrow(r2+offsetx, 0, 20,0,width=1,color='r')
ax.text(r2+offsetx+25,0,f'Fr4')

ax.annotate("", xy=(r2, 10), xytext=(r2, -10),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
ax.annotate("", xy=(r2+offsetx, -10), xytext=(r2+offsetx, 10),
            arrowprops=dict(arrowstyle="<-",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
plt.show()
```

```python
n3=n2
print(f'n3={n3:.0f}rpm')
Torque3=round(Power*60/(2*pi*n3),2)
print(f'Torque={Torque3} Nm')
```

Find radius for Gear 3
$$r_3=\frac{0.08}{2}$$
$$r_3=0.04m$$

Use torque equation

$$T=F_{t3}\times r_3$$
$$14.32=F_{t3}(0.04)$$

```python
Ft3 = round(Torque3/r3,2)
print(f'Ft3={Ft3} N')
```

Radial force

$$F_{r3}=F_{t3} \tan(20)$$
$$=(358)\tan(20)$$

```python
Fr3=round(Ft3*tan(pressureAngle),2)
print(f'Fr3={Fr3} N')
```

$$F_{t3}=F_{t4}$$

iii) If the diameter of gear 4 is increased to 120mm, explain how it affects the torque of gear 4

Current speed

```python
print(f'Speed of Gear 4={n4:.0f} rpm')
print(Power)
Torque4=round(Power*60/(2*pi*n4),2)
print(f'Torque={Torque4} Nm')
```

$$d_4=100mm \rightarrow d'_4=120mm$$

$$\frac{n_3}{n_4}=\frac{d_4}{d_3}$$
$$\frac{1200}{n_4}=\frac{120}{80}$$

```python
new_d4=120
new_r4=new_d4/2
print(f'New r4={new_r4} mm')
new_n4=n3*d3/new_d4
print(f'New n4={new_n4} rpm')
```

```python
# Draw Gear
fig, ax = plt.subplots()

offsetx=10
offsety=5

r3=40

circle1 = plt.Circle((0, 0), r3, color='green', fill=False)  
circle2 = plt.Circle((r3+new_r4, 0), new_r4, color='brown', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')
ax.axis('off')

ax.set_xlim(-60, 180)
ax.set_ylim(-80, 80)

# ax.set(title='Gear 1 and 2')
ax.text(-offsetx,r3+offsety,f'Gear 3')
ax.text(-offsetx+r3+new_r4,new_r4+offsety,f'New Gear 4')
ax.text(-offsetx,-r3-offsety,f'd3={d3}mm')
ax.text(r3+new_r4,-new_r4-offsety,f'd4={new_d4}mm')
plt.show()
```

From torque equation

$$P=\frac{2\pi n_3}{60} T'$$
$$1800=\frac{2\pi (800)}{60} T'$$
$$T'=21.49N.m$$

```python
Torque4=round(Power*60/(2*pi*new_n4),2)
print(f'New Torque at Gear 4={Torque4}Nm')
print('When diameter of Gear 4 increase, the transmitted torque increase')
```
