ii) Determine tangential and radial forces between Gear 1 and 2 and between gear 3 and 4.

From torque equation

$$P=\frac{2\pi n_1}{60} T$$
$$3000=\frac{2\pi (2000)}{60} T$$

```python
Torque=round(Power*60/(2*pi*n1),2)
print(f'Torque={Torque} Nm')
```

From module equation $$m=\frac{d_1}{N_1}$$
$$8=\frac{d_1}{200}$$


```python
d1=module*N1
r1=d1/2/1000
print(f'd1={d1} mm')
print(f'r1={r1} m')
```

Use torque equation

$$T=F_{t1}\times r_1$$
$$14.32=F_{t1}(0.8)$$

```python
Ft1=Torque/r1
print(f'Tangential Force, Ft1={Ft1} N')
```

Radial force

$$F_{r1}=F_{t1} \tan(20)$$
$$=(17.9)\tan(20)$$

```python
Fr1=round(Ft1*tan(pressureAngle),2)
print(f'Fr1={Fr1}N')
```

```python
# FBD between Gear 1 and 2
fig, ax = plt.subplots()

circle1 = plt.Circle((0, 0), 16, color='blue', fill=False)  
circle2 = plt.Circle((65, 0), 40, color='blue', fill=False) 

ax.add_patch(circle1)
ax.add_patch(circle2)

ax.set_aspect('equal')

ax.set_xlim(-20, 120)
ax.set_ylim(-50, 50)

ax.set(title='FBD between Gear 1 and 2')
ax.text(0,25,f'Gear 1')
ax.text(56,40,f'Gear 2')
ax.text(-15,-18,f'n1=2000 rpm')
ax.text(46,-10,f'n2=800 rpm')


ax.arrow(16, 0, 0,-40,width=1,color='r')
ax.text(5,-40,f'Ft1')
ax.arrow(25, 0, 0,40,width=1,color='r')
ax.text(15,40,f'Ft2')

ax.arrow(16, 0, -10,0,width=1,color='r')
ax.text(-5,0,f'Fr1')
ax.arrow(25, 0, 6,0,width=1,color='r')
ax.text(40,0,f'Fr2')

ax.annotate("", xy=(0, 20), xytext=(15, 10),
            arrowprops=dict(arrowstyle="<-",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
ax.annotate("", xy=(20, 20), xytext=(30, 30),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=0.2",
                            color="green", lw=2))
plt.show()
```

iii) If number of teeth for gear 4 decreased to 8, what happen to gear 4

$$N_4=12 \rightarrow N'_4=8$$

```python
d4=module*N4
r4=d4/2/1000
print(f'Number of teeth for Gear 4: {N4}')
print(f'd4={d4} mm')
print(f'r4={r4} m')
```

$$T_4=F_{t4}\times r_4$$
$$T_4=(746.04)(0.12)$$

```python
Ft4=Ft3
Torque4=Ft4*r4
print(f'Torque at Gear 4={Torque4}Nm')
```

New Torque at Gear 4

Number of teeth changed to 4, the diameter will changed.

$$8=\frac{d'_4}{8}$$

```python
N4_new=8
d4_new=module*N4_new
r4_new=d4_new/2/1000
print(f'Number of teeth for Gear 4: {N4_new}')
print(f'd4={d4_new} mm')
print(f'r4={r4_new} m')
```

```python
Torque4_new=round(Ft4*r4_new,2)
print(f'New Torque at Gear 4={Torque4_new}Nm')
print('When number of teeth decrease, the transmitted torque also decrease')
```
